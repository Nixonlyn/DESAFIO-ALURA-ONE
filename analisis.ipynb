# Análisis de ventas
Este proyecto realiza un análisis exploratorio de datos sobre un conjunto de ventas.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('ventas.csv')
df.head()
```

```
df['Total'] = df['Precio'] * df['Cantidad']
ventas_categoria = df.groupby('Categoria')['Total'].sum()
ventas_categoria.plot(kind='bar')
plt.show()
```
